```python
#clean the agricultural data 
# Remove duplicates
df.drop_duplicates(inplace=True)

# Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

# Handle missing values
df.fillna(0, inplace=True)

# Trim text columns
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip()
```

```python
import pandas as pd
import mysql.connector
import numpy as np
import os

# -------------------------------
# STEP 1: LOAD CSV
# -------------------------------
file_path = r"C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\Project_Agri_2_GUVI\agri_cleaned_data.csv"

df = pd.read_csv(file_path)

# Replace NaN with None (important for MySQL)
df = df.replace({np.nan: None})

print("âœ… CSV loaded")
print("Rows:", len(df))
print("Columns:", len(df.columns))

# -------------------------------
# STEP 2: MYSQL CONNECTION
# -------------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Adlin@2025",
    database="agriculture_db"
)
cursor = conn.cursor()

# -------------------------------
# STEP 3: PREPARE INSERT QUERY
# -------------------------------
table_name = "crop_data"

columns = ", ".join(f"`{col}`" for col in df.columns)
placeholders = ", ".join(["%s"] * len(df.columns))

insert_query = f"""
INSERT INTO {table_name} ({columns})
VALUES ({placeholders})
"""

# -------------------------------
# STEP 4: INSERT ALL RECORDS
# -------------------------------
data = [tuple(row) for row in df.itertuples(index=False, name=None)]

cursor.executemany(insert_query, data)
conn.commit()

print(f"âœ… {cursor.rowcount} records inserted successfully")

# -------------------------------
# STEP 5: CLOSE CONNECTION
# -------------------------------
cursor.close()
conn.close()

print("ðŸŽ‰ DATA INSERT COMPLETED")
```
